```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('adult_data.csv')
df.columns = ['age','workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation',
             'relationship', 'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'salary']
```

```python
df.head()
```

```python
df.info()
```

```python
df['salary'].value_counts()
```

```python
df['salary'] = np.where(df['salary'].str.endswith('<=50K'),0,1)
```

```python
df.drop(['fnlwgt'],1,inplace=True)
```

```python
dummy = pd.get_dummies(df[['workclass','education','marital_status','occupation','relationship','race','sex','native_country']],drop_first=True)
```

```python
df.drop(['workclass','education','marital_status','occupation','relationship','race','sex','native_country'],1,inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
df.iloc[:,:-1]=sc.fit_transform(df.iloc[:,:-1])
```

```python
df = pd.concat([df,dummy],1)
```

```python
df
```

```python
X = df.drop(['salary'],1)
y = df.salary
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
```

```python
LR = LogisticRegression()
DT = DecisionTreeClassifier()
RF = RandomForestClassifier()
SV = SVC()
KNN = KNeighborsClassifier()
XG = XGBClassifier()
```

```python
from sklearn.metrics import accuracy_score, f1_score
```

```python
models = [LR,DT,RF,SV,KNN,XG]
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(model)    
    print('Training Accuracy Score: ', model.score(X_train, y_train))
    print('Testing Accuracy Score: ', accuracy_score(y_test, y_pred))
    print('F1_Score: ', f1_score(y_test,y_pred))
    print()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score_rf = cross_val_score(XGBClassifier(),X,y,cv=5)
score_rf
```
